```python
def factor_remover(data, col_rm=None, row_rm=None):
    if col_rm:
        data = data.drop(columns=col_rm)

    if row_rm:
        data = data.drop(row_rm)

    return data

rm = data[data.DefaultData == True].index.tolist()
rm1 = data[data["ComplianceStatus"] == "Missing Data"].index.tolist()

# Combine rm and rm1 into a single list
row_rm = rm + rm1

cleaned_data = factor_remover(data, col_rm=["OSEBuildingID","DataYear","City","State","Comments"], row_rm=row_rm)
cleaned_data
```
